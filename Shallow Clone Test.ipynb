# 🚀 Supercharge Your Data Workflow with Shallow Clones!
Discover how Shallow Clones let you quickly create lightweight copies of your tables for faster experimentation and collaboration.

## Create data

```python
sql("CREATE CATALOG IF NOT EXISTS cw_dbx_testing")

sql("CREATE SCHEMA IF NOT EXISTS cw_dbx_testing.bookstore")
```

```python
# Import necessary types for schema definition and date handling
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType, DateType
from datetime import date

# Sample data representing orders
data = [
    (1, 101, 'Completed', 29.99, date(2026, 3, 1), 'High', 'Clerk#0001', 0, 'First order'),
    (2, 102, 'Pending', 15.50, date(2026, 3, 2), 'Medium', 'Clerk#0002', 0, 'Awaiting payment'),
    (3, 103, 'Completed', 45.00, date(2026, 3, 3), 'Low', 'Clerk#0003', 0, 'Gift order'),
    (4, 104, 'Cancelled', 12.99, date(2026, 3, 4), 'High', 'Clerk#0004', 0, 'Customer cancelled'),
    (5, 105, 'Completed', 22.75, date(2026, 3, 5), 'Medium', 'Clerk#0005', 0, 'Repeat customer'),
    (6, 106, 'Pending', 18.20, date(2026, 3, 6), 'Low', 'Clerk#0006', 0, 'Awaiting shipment'),
    (7, 107, 'Completed', 35.99, date(2026, 3, 7), 'High', 'Clerk#0007', 0, 'Bulk order'),
    (8, 108, 'Completed', 27.50, date(2026, 3, 8), 'Medium', 'Clerk#0008', 0, 'Express delivery'),
    (9, 109, 'Pending', 16.80, date(2026, 3, 9), 'Low', 'Clerk#0009', 0, 'New customer'),
    (10, 110, 'Completed', 40.00, date(2026, 3, 10), 'High', 'Clerk#0010', 0, 'Special request')
]

# Define the schema for the orders data
schema = StructType([
    StructField("o_orderkey", IntegerType(), False),
    StructField("o_custkey", IntegerType(), False),
    StructField("o_orderstatus", StringType(), False),
    StructField("o_totalprice", DoubleType(), False),
    StructField("o_orderdate", DateType(), False),
    StructField("o_orderpriority", StringType(), False),
    StructField("o_clerk", StringType(), False),
    StructField("o_shippriority", IntegerType(), False),
    StructField("o_comment", StringType(), False)
])

# Create a Spark DataFrame using the sample data and schema
orders_df = spark.createDataFrame(data, schema)

# Define the full table name including catalog and schema
table_fullname = "cw_dbx_testing.bookstore.Orders"

# Write the DataFrame to a Delta table, overwriting if it already exists
orders_df.write.mode("overwrite").saveAsTable(table_fullname)

# Display the contents of the newly created table
display(spark.table(table_fullname))
```

## Create Shallow code for testing

```python
sql("""
CREATE TABLE cw_dbx_testing.bookstore.CustomerOrders
SHALLOW CLONE cw_dbx_testing.bookstore.Orders
""")
```

```python
display(spark.table("cw_dbx_testing.bookstore.CustomerOrders").count())
```

## Compare source table and shallow clone table for row counts

```python
# Get the row count of the Orders table
orders_count = spark.table("cw_dbx_testing.bookstore.Orders").count()

# Get the row count of the CustomerOrders table
customerorders_count = spark.table("cw_dbx_testing.bookstore.CustomerOrders").count()

# Determine if the row counts match
match_status = "Match" if orders_count == customerorders_count else "Mismatch"

# Display the counts and validation status in a DataFrame
display(spark.createDataFrame(
    [(orders_count, customerorders_count, match_status)],
    ["Orders_count", "CustomerOrders_count", "RowCount_Validation"]
))
```
